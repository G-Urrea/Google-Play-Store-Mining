# Pregunta 2: ¿Es posible caracterizar tipos de aplicaciones respecto a los reviews recibidos por los usuarios? 

## Propuesta metodológica

Sobre los datos de 'Apps/Reviews' se aplican los siguentes métodos: _K_-means, con valores de _k_ obtenidos según optimos observables en un gráfico de codo; clustering jerárquico aglomerativo, con criterios _complete_, _average_, _single_ y _ward_; y DBSCAN con valores de _eps_ y _minPts_ dados por la distancia media a los _k_ vecinos y el doble de la dimensionalidad del dataset respectivamente, ambos valores se toman como base y se variarán según los resultados.

Se tendrán dos enfoques para evaluar los clusters. La primera se centra en los propios datos usados para crear los clusters, es decir el de las reviews. Los resultados se evaluan en base a cohesión (SSE) y coeficiente de Silhouette. Para visualizar los resultados se reduce la dimensionalidad con el método _principal component analisys_, de modo que sea posible crear un gráfico de disperción. También se estudia la distribución de los atributos en los clusters resultantes, en específico las medias y cuartiles.

Para este caso se podrían aplicar más filtros y dividir los datos en subconjuntos más homogéneos (según número de descargas, categorías, content rating, ect.) sin embargo, el dataset actual ya es demasiado pequeño (607 filas).

La segunda opción es evaluar la posibilidad de que, por medio de sus reviews, se logre caracterizar las aplicaciones según datos externos a esta, como lo son la categoría o el rating. En este caso se evaluan los resultados comprobando las diferencias y similitudes de las aplicaciones de los clusters resultantes, según sus atributos en el dataframe 'Apps'. Para ello se considera que fracción de categorías, géneros y content rating se encuentran en cada cluster, junto con el porcentanje de la clase mayoritaria en cada caso. Se considerará también la media de rating de cada cluster. Estos resultados también pueden ser útiles para responder la pregunta 3.

Para ambos casos se evalua también con medida de información mutua, comparando los clusters con distintos atributos del dataset.

```python
def reEscale(data, a, b):
    m = min(data)
    M = max(data)
    p = (b-a)/(M-m)
    c = (a*M-b*m)/(M-m)
    data2 = np.zeros(len(data))
    for i in range(len(data)):
        data2[i] = p*data[i]+c
    return data2.copy()
```

```python
App_Review2["chars_excl_spaces_count"] = reEscale(np.array(App_Review2["chars_excl_spaces_count"]), 0, 1)
App_Review2["duplicates_count"] = reEscale(np.array(App_Review2["duplicates_count"]), 0, 1)
```

## Dataset para aplicar cluster

Debido a que la mayoría de los datos del texto presentan alta correlación, se reduce su dimensionalidad a 2 (arbitrario).

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import make_pipeline
```

```python
svd = TruncatedSVD(2)
#normalizer = Normalizer(copy=False)
#lsa = make_pipeline(svd, normalizer)
redCount = svd.fit_transform(App_Review2.iloc[:,19:28])
```

```python
X = pd.DataFrame()
X["Reviews"] = reEscale(np.array(App_Review2["Translated_Review"]), 0, 1)
X["Polarity"] = reEscale(np.array(App_Review2["sentiment_polarity_score"]), 0, 1)
X["Subjectivity"] = reEscale(np.array(App_Review2["sentiment_subjectivity_score"]), 0, 1)
X["ease_of_reading"] = reEscale(np.array(App_Review2["ease_of_reading_score"]), 0, 1)
X["spelling"] = reEscale(np.array(App_Review2["spelling_quality_score"]), 0, 1)
X["textData1"] = reEscale(redCount.transpose()[0], 0, 1)
X["textData2"] = reEscale(redCount.transpose()[1], 0, 1)
```

### K-Means

```python
sse = []

clusters = list(range(1, 30))
for k in clusters:
    kmeans = KMeans(n_clusters=k).fit(X)
    sse.append(kmeans.inertia_)

plt.plot(clusters, sse, marker="o")
plt.title("Método del codo de 1 a 30 clusters")
plt.grid(True)
plt.show()
```

```python
def ploterK5(dt, labels, title):
    n = len(labels)
    m = np.arange(1,1+2*n,2)
    width = 0.3
    
    fig, ax = plt.subplots(figsize=(3+n,4))
    cp1 = dict()
    cp1[0] = ax.boxplot(dt[dt.label==0].iloc[:, 0:n], positions=m-width*2, widths=width, patch_artist=True)
    cp1[1] = ax.boxplot(dt[dt.label==1].iloc[:, 0:n], positions=m-width, widths=width, patch_artist=True)
    cp1[2] = ax.boxplot(dt[dt.label==2].iloc[:, 0:n], positions=m, widths=width, patch_artist=True)
    cp1[3] = ax.boxplot(dt[dt.label==3].iloc[:, 0:n], positions=m+width, widths=width, patch_artist=True)
    cp1[4] = ax.boxplot(dt[dt.label==4].iloc[:, 0:n], positions=m+width*2, widths=width, patch_artist=True)
    
    colors = ["red", "blue", "green", "purple", "orange"]
    for i in range(5):
        for box in cp1[i]['boxes']:
            box.set_facecolor(colors[i])
    
    ax.set_title(title)
    ax.set_xticks(m)
    ax.set_xticklabels(labels)
    
    fig.tight_layout()
    
    plt.show()
```

```python
Y1 = X.iloc[:, 0:4].copy()
Y1["label"] = km[0].labels_

ploterK5(Y1, ["Polarity", "Subjectivity", "Ease of Reading", "Spelling"], "Scores por cluster")
```

```python
Y2 = App_Review2.iloc[:, 19:24].copy()
Y2["Review"] = X["Reviews"].copy()
Y2["label"] = km[0].labels_

ploterK5(Y2, ["Sentences", "Spaces", "Words", "Characters", "Duplicated", "Reviews"], "Conteos por cluster")
```

```python
Y3 = App_Review2.iloc[:, 7:11].copy()
Y3["label"] = km[0].labels_
Y3 = Y3.dropna(how='any')
Y3 = Y3[Y3.Installs < 50000000]
Y3 = Y3[Y3.Reviews < 1500000]
Y3["Rating"] = reEscale(np.array(Y3.Rating), 0, 1)
Y3["Reviews"] = reEscale(np.array(Y3.Reviews), 0, 1)
Y3["Size"] = reEscale(np.array(Y3.Size), 0, 1)
Y3["Installs"] = reEscale(np.array(Y3.Installs), 0, 1)

ploterK5(Y3, ["Rating", "Reviews", "Size", "Installs"], "Datos de App por cluster")
```

```python
def ploterCom(dt, labels, title):
    n = len(labels)
    m = np.arange(1,1+2*n,2)
    width = 0.3
    
    fig, ax = plt.subplots(figsize=(3+n,4))
    cp1 = dict()
    cp1[0] = ax.boxplot(dt[dt.label==0].iloc[:, 0:n], positions=m-width*2.5, widths=width, patch_artist=True)
    cp1[1] = ax.boxplot(dt[dt.label==1].iloc[:, 0:n], positions=m-width*1.5, widths=width, patch_artist=True)
    cp1[2] = ax.boxplot(dt[dt.label==2].iloc[:, 0:n], positions=m-0.5*width, widths=width, patch_artist=True)
    cp1[3] = ax.boxplot(dt[dt.label==3].iloc[:, 0:n], positions=m+width*0.5, widths=width, patch_artist=True)
    cp1[4] = ax.boxplot(dt[dt.label==4].iloc[:, 0:n], positions=m+width*1.5, widths=width, patch_artist=True)
    cp1[5] = ax.boxplot(dt[dt.label==4].iloc[:, 0:n], positions=m+width*2.5, widths=width, patch_artist=True)
    
    colors = ["red", "blue", "green", "yellow", "purple", "orange"]
    for i in range(6):
        for box in cp1[i]['boxes']:
            box.set_facecolor(colors[i])
    
    ax.set_title(title)
    ax.set_xticks(m)
    ax.set_xticklabels(labels)
    
    fig.tight_layout()
    
    plt.show()
```
